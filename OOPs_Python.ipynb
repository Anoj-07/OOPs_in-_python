# Classes and Objects

```python
# what is class?
# => A blue print for creating object that defines attributes(data) and methods(behaviours)

class Person:
  # attribute
  species = "Human" # global variable

  # method
  def greet(self):
    print(f"hello, I am {self.species}")
```

```python
# what is object?
# => An instance of a class which Has its own copy of attributes and can use class methods.

person1 = Person() # object creation
person1.greet() # Method call -> Hello, I am a person
print(person1.species) # Acess class attributes -> Human
```
